```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fftandplot(x,y): # define the function to execute Fourier transform and then plot all the functions/spectra.
    
    # Fourier transform
    y_fft = np.fft.fft(y,int(N+1))*dt # y_fft multiplied by dt to correct for the amplitude.
    x_fft = np.fft.fftfreq(int(N+1),dt)

    # Plot original function
    fig, ax = plt.subplots()
    plt.title('Original Function')
    plt.xlabel('t (sec)')
    plt.ylabel('y(t)')
    ax.plot(x,y.real,'b')
    ax.plot(x,y.imag,'r')

    # Plot FT real part
    fig, ax = plt.subplots()
    plt.title('Fourier Transform (real part)')
    plt.xlabel('f (Hz)')
    plt.ylabel('Y(f)')
    ax.plot(x_fft,y_fft.real,'b')
    plt.show()

    # Plot FT real part (zoomed in)
    fig, ax = plt.subplots()
    plt.title('Fourier Transform (real part zoomed in)')
    plt.xlabel('f (Hz)')
    plt.ylabel('Y(f)')
    ax.plot(x_fft[0:int(N/8)],y_fft[0:int(N/8)].real,'b')
    ax.plot(x_fft[int(7*N/8):(N+1)],y_fft[int(7*N/8):(N+1)].real,'b')
    plt.show()
    
    # Plot FT imaginary part
    fig, ax = plt.subplots()
    plt.title('Fourier Transform (imaginary part)')
    plt.xlabel('f (Hz)')
    plt.ylabel('Y(f)')
    ax.plot(x_fft,y_fft.imag,'r')
    plt.show()

    # Plot FT imaginary part (zoomed in)
    fig, ax = plt.subplots()
    plt.title('Fourier Transform (imaginary part zoomed in)')
    plt.xlabel('f (Hz)')
    plt.ylabel('Y(f)')
    ax.plot(x_fft[0:int(N/8)],y_fft[0:int(N/8)].imag,'r')
    ax.plot(x_fft[int(7*N/8):(N+1)],y_fft[int(7*N/8):(N+1)].imag,'r')
    plt.show()

    # Amplitude spectrum
    fig, ax = plt.subplots()
    plt.title('Amplitude Spectrum')
    plt.xlabel('f (Hz)')
    plt.ylabel('Amplitude')
    ax.plot(x_fft,np.abs(y_fft),'g')
    plt.show()
    
    # Amplitude spectrum (zoomed in)
    fig, ax = plt.subplots()
    plt.title('Amplitude Spectrum (zoomed in)')
    plt.xlabel('f (Hz)')
    plt.ylabel('Amplitude')
    ax.plot(x_fft[0:int(N/8)],np.abs(y_fft[0:int(N/8)]),'g')
    ax.plot(x_fft[int(7*N/8):(N+1)],np.abs(y_fft[int(7*N/8):(N+1)]),'g')
    plt.show()


    # Phase spectrum
    fig, ax = plt.subplots()
    plt.title('Phase Spectrum')
    plt.xlabel('f (Hz)')
    plt.ylabel('Phase')
    ax.plot(x_fft,np.angle(y_fft),'y')
    plt.show()
 
    # Phase spectrum (zoomed in)
    fig, ax = plt.subplots()
    plt.title('Phase Spectrum (zoomed in)')
    plt.xlabel('f (Hz)')
    plt.ylabel('Phase')
    ax.plot(x_fft[0:int(N/8)],np.angle(y_fft[0:int(N/8)]),'y')
    ax.plot(x_fft[int(7*N/8):(N+1)],np.angle(y_fft[int(7*N/8):(N+1)]),'y')
    plt.show()
```

```python
# Total time (sec)
T = 10.24
# Time interval (sec)
dt = 0.01
# Number of samplepoints
N = int(T/dt)

# Amplitude
B = 3
# Frequency (Hz)
fy = 3.125
```

### Question 1

```python
x = np.linspace(0, T, num=int(N+1))
y = B*np.cos(2*np.pi*fy*x)
fftandplot(x,y)
```

### Question 2

```python
t_p = 0.2

x = np.linspace(0, T, num=int(N+1))
y = B*np.cos(2*np.pi*fy*(x-t_p))
fftandplot(x,y)
```

### Question 3

```python
t_p = 3
w = 2

x = np.linspace(0, T, num=int(N+1))
y = np.linspace(0, 0, num=int(N+1))
y[int((t_p-w/2)/dt):int((t_p+w/2)/dt+1)] = B

fftandplot(x,y)
```

### Question 4

```python
t_p = 4
w = 2

x = np.linspace(0, T, num=int(N+1))
y = np.linspace(0, 0, num=int(N+1))
y[int((t_p-w)/dt):int(t_p/dt)] = -B
y[int(t_p/dt):int((t_p+w)/dt+1)] = B

fftandplot(x,y)
```

### Question 5

```python
t_p = 4
w = 2

x = np.linspace(0, T, num=int(N+1))
y = np.linspace(0, 0, num=int(N+1))
y[int((t_p-w)/dt):int(t_p/dt)] = B/w*(w+x[int((t_p-w)/dt):int(t_p/dt)]-t_p)
y[int(t_p/dt):int((t_p+w)/dt+1)] = B/w*(w-x[int(t_p/dt):int((t_p+w)/dt+1)]+t_p)

fftandplot(x,y)
```

### Question 6

```python
t_p = T/2
sigma = 1

x = np.linspace(0, T, num=int(N+1))
y = B*np.exp(-(x-t_p)**2/(2*sigma**2))

fftandplot(x,y)
```
