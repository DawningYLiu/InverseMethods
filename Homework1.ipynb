```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Total time (sec)
T = 10.24
# Time interval (sec)
dt = 0.01
# Number of samplepoints
N = T/dt

# Amplitude
B = 3
# Frequency (Hz)
fy = 2.56

x = np.linspace(0, T, num=int(N+1))
y = B*np.cos(2*np.pi*fy*x + np.pi/4)
y_fft = np.fft.fft(y,int(N+1))/(N+1)
x_fft = np.fft.fftfreq(int(N+1),dt)

fig, ax = plt.subplots()
plt.title('Original Function')
plt.xlabel('t (sec)')
plt.ylabel('y(t)')
ax.plot(x,y.real,'b')
ax.plot(x,y.imag,'r')

fig, ax = plt.subplots()
plt.title('Fourier Transform (real part)')
plt.xlabel('f (Hz)')
plt.ylabel('Y(f)')
ax.plot(x_fft,y_fft.real,'b')
plt.show()

fig, ax = plt.subplots()
plt.title('Fourier Transform (imaginary part)')
plt.xlabel('f (Hz)')
plt.ylabel('Y(f)')
ax.plot(x_fft,y_fft.imag,'r')
plt.show()

# Amplitude spectrum
fig, ax = plt.subplots()
plt.title('Amplitude Spectrum')
plt.xlabel('f (Hz)')
plt.ylabel('Amplitude')
ax.plot(x_fft,np.abs(y_fft),'g')
plt.show()

# Phase spectrum
fig, ax = plt.subplots()
plt.title('Phase Spectrum')
plt.xlabel('f (Hz)')
plt.ylabel('Phase')
ax.plot(x_fft,np.angle(y_fft))
plt.show()
```
